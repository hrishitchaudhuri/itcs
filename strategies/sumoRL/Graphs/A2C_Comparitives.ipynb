```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ncpu1 = pd.read_csv('../outputs/a2c_22.5k/Test_uniform/uniform_run1.csv')
ncpu1_x = ncpu1.step_time
ncpu1_y = ncpu1.total_wait_time
```

```python
plt.plot(ncpu1_x, ncpu1_y)
plt.xlabel('Time Step')
plt.ylabel('Waiting Times')
```

```python
#ncpu1 = pd.read_csv('../outputs/a2c_22.5k/Test_uniform/uniform_run1.csv')
ncpu2 = pd.read_csv('../outputs/a2c_ncpu=2/Test/uniform_new_run1.csv')
ncpu2_x = ncpu2.step_time
ncpu2_y = ncpu2.total_wait_time
```

```python
plt.plot(ncpu2_x, ncpu2_y)
plt.xlabel('Time Step')
plt.ylabel('Wait Times')
```

```python
fig, ax = plt.subplots()
ax.plot(ncpu1_x, ncpu1_y, '-b', label="N CPU = 1")
ax.plot(ncpu2_x, ncpu2_y, '-r', label="N CPU = 2")
ax.legend()
plt.xlabel('Time Step')
plt.ylabel('Waiting Times')
#plt.title('Training run vs Testing run')
fig = plt.gcf()
fig.set_size_inches(10, 6)
fig.savefig('A2C_comp/ncpu=1vs2', dpi=96)
```

```python
ncpu4 = pd.read_csv('../outputs/a2c_ncpu=4/Test_45k/uniform_run1.csv')
ncpu4_x = ncpu4.step_time
ncpu4_y = ncpu4.total_wait_time
```

```python
fig, ax = plt.subplots()
ax.plot(ncpu1_x, ncpu1_y, '-b', label="N CPU = 1")
ax.plot(ncpu4_x, ncpu4_y, '-g', label="N CPU = 4")
ax.legend()
plt.xlabel('Time Step')
plt.ylabel('Waiting Times')
#plt.title('Training run vs Testing run')
fig = plt.gcf()
fig.set_size_inches(10, 6)
fig.savefig('A2C_comp/ncpu=1vs4', dpi=96)
```

```python
ncpu1_wait_time = sum(ncpu1_y)
ncpu2_wait_time = sum(ncpu2_y)
ncpu4_wait_time = sum(ncpu4_y)
wait_times = [ncpu1_wait_time, ncpu2_wait_time, ncpu4_wait_time]
plt.bar(['1', '2', '4'], wait_times, color=['blue', 'red', 'green'])
plt.xlabel('Number of Processes')
plt.ylabel('Total wait time (over full simulation)')
fig = plt.gcf()
#fig.set_size_inches(20, 11.25)
fig.savefig('A2C_comp/wait_time_bar', dpi=96)
```

```python
ncpu1_train_time = 515.198
ncpu2_train_time = 344.861
ncpu4_train_time = 258.531
train_times = [ncpu1_train_time, ncpu2_train_time, ncpu4_train_time]
plt.bar(['1', '2', '4'], train_times, color=['blue', 'red', 'green'])
plt.xlabel('Number of Processes')
plt.ylabel('Training time (in seconds)')
fig = plt.gcf()
fig.savefig('A2C_comp/train_time_bar', dpi=96)
```
