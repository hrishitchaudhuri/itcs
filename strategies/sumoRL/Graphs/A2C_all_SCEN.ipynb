```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
uniform = pd.read_csv('../outputs/a2c_22.5k/Test_uniform/uniform_run1.csv')
uni_x = uniform.step_time
uni_y = uniform.total_stopped
```

```python
plt.plot(uni_x, uni_y)
plt.xlabel('Time Step')
plt.ylabel('Queue Length (vehicles)')
fig = plt.gcf()
fig.set_size_inches(20, 11.25)
fig.savefig('A2C_all_SCEN/uniform', dpi=96)
```
