```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../Results 5400 seconds/dqn_final/train_run1.csv')
test = pd.read_csv('../Results 5400 seconds/dqn_final/train_run5.csv')
```

```python
train_x = train.step_time
train_y = train.total_wait_time
test_x = test.step_time
test_y = test.total_wait_time
```

```python
plt.plot(train_x, train_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Training phase')
```

```python
plt.plot(test_x, test_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Testing phase')
```

```python
fig, ax = plt.subplots()
ax.plot(train_x, train_y, '-b', label="Training run")
ax.plot(test_x, test_y, '-r', label="Testing run")
ax.legend()
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Training run vs Testing run')
```

```python
train_wait_time = sum(train_y)
test_wait_time = sum(test_y)
times = [train_wait_time, test_wait_time]
plt.bar(['Train', 'Test'], times)
plt.ylabel('Total wait time (across full simulation)')
```
