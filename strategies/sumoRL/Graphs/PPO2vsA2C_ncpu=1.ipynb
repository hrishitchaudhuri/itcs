```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a2c = pd.read_csv('a2c_final/train_run13.csv')
ppo = pd.read_csv('ppo2_final/train_run13.csv')
```

```python
a2c_x = a2c.step_time
a2c_y = a2c.total_wait_time
ppo_x = ppo.step_time
ppo_y = ppo.total_wait_time
```

```python
plt.plot(a2c_x, a2c_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('A2C')
```

```python
plt.plot(ppo_x, ppo_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('PPO')
```

```python
fig, ax = plt.subplots()
ax.plot(a2c_x, a2c_y, '-b', label="A2C")
ax.plot(ppo_x, ppo_y, '-r', label="PPO")
ax.legend()
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('A2C vs PPO')
```

```python
a2c_wait_time = sum(a2c_y)
ppo_wait_time = sum(ppo_y)
times = [a2c_wait_time, ppo_wait_time]
plt.bar(['A2C', 'PPO'], times)
plt.ylabel('Total wait time (across full simulation)')
```
