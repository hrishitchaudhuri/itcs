```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ncpu_1 = pd.read_csv('a2c_final/train_run13.csv')
ncpu_4 = pd.read_csv('a2c_final_ncpu=4_2ndtime/train_run4.csv')
```

```python
ncpu1_x = ncpu_1.step_time
ncpu1_y = ncpu_1.total_wait_time
ncpu4_x = ncpu_4.step_time
ncpu4_y = ncpu_4.total_wait_time
```

```python
plt.plot(ncpu1_x, ncpu1_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Number of workers = 1')
```

```python
plt.plot(ncpu4_x, ncpu4_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Number of workers = 4')
```

```python
fig, ax = plt.subplots()
ax.plot(ncpu1_x, ncpu1_y, '-b', label="N CPU = 1")
ax.plot(ncpu4_x, ncpu4_y, '-r', label="N CPU = 4")
ax.legend()
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('No of workers = 1 vs No of workers = 4')
```

```python
ncpu1_wait_time = sum(ncpu1_y)
ncpu4_wait_time = sum(ncpu4_y)
times = [ncpu1_wait_time, ncpu4_wait_time]
plt.bar(['NCPU = 1', 'NCPU = 4'], times)
plt.ylabel('Total wait time (across full simulation)')
```
