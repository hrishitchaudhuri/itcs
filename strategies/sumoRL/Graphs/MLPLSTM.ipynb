```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('a2c_final_Lstm/train_run1.csv')
test = pd.read_csv('a2c_final_Lstm/train_run13.csv')
```

```python
train_x = train.step_time
train_y = train.total_wait_time
test_x = test.step_time
test_y = test.total_wait_time
```

```python
plt.plot(train_x, train_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Training phase')
```

```python
plt.plot(test_x, test_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Testing phase')
```

```python
fig, ax = plt.subplots()
ax.plot(train_x, train_y, '-b', label="Training run")
ax.plot(test_x, test_y, '-r', label="Testing run")
ax.legend()
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Training run vs Testing run')
```
