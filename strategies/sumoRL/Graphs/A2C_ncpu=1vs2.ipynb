```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ncpu_1 = pd.read_csv('a2c_final/train_run13.csv')
ncpu_2 = pd.read_csv('a2c_final_ncpu=2/train_run10.csv')
```

```python
ncpu1_x = ncpu_1.step_time
ncpu1_y = ncpu_1.total_wait_time
ncpu2_x = ncpu_2.step_time
ncpu2_y = ncpu_2.total_wait_time
```

```python
plt.plot(ncpu1_x, ncpu1_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Number of workers = 1')
```

```python
plt.plot(ncpu2_x, ncpu2_y)
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('Number of workers = 2')
```

```python
fig, ax = plt.subplots()
ax.plot(ncpu1_x, ncpu1_y, '-b', label="N CPU = 1")
ax.plot(ncpu2_x, ncpu2_y, '-r', label="N CPU = 2")
ax.legend()
plt.xlabel('Time Step')
plt.ylabel('Total Wait Time')
plt.title('No of workers = 1 vs No of workers = 2')
```

```python
ncpu1_wait_time = sum(ncpu1_y)
ncpu2_wait_time = sum(ncpu2_y)
times = [ncpu1_wait_time, ncpu2_wait_time]
plt.bar(['NCPU = 1', 'NCPU = 2'], times)
plt.ylabel('Total wait time (across full simulation)')
```
